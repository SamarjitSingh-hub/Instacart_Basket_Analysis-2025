# Data Consistency Checks

### Mixed Type Data


```python
# Creating a new Data frame with mixed datatype column named df_test
df_test = pd.DataFrame()
```

```python
# Imputing mixed data type values in the new dataframe
df_test['mix'] = ['a','b','1','True']
```

```python
df_test
```

```python
#Creating fuction to search for a mixed data type in a column
for col in df_test.columns.tolist():
  weird = (df_test[[col]].applymap(type) != df_test[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_test[weird]) > 0:
    print (col)
```

```python
#Converting the mixed column data type to String.
df_test['mix'] = df_test['mix'].astype('str')
```

```python
df_test['mix'].dtype
```

dtype 0 above confirms that the columns in the df_test have been successfully coverted to string
