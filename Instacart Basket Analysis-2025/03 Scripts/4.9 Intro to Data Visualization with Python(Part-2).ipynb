### Check whether there’s a difference in expenditure (the “prices” column) depending on the hour of the day

```python
# Generating a sample of the dataframe
sampled_df = df.sample(n=100000, random_state=1)
```

```python
# GROUP BY the hour of the day and calculate the mean expenditure
hourly_expenditure = sampled_df.groupby('order_hour_of_day')['prices'].mean().reset_index()
```
