```python
# Importing Dataframe
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_merge_updated.pkl'))
```

```python
# Creating a subset
ords_prods_merge_subset = ords_prods_merge[:1000000]
```

```python
# Performing aggregation function on the subset of the dataframe
ords_prods_merge_subset.groupby(['department_id']).agg({'order_number':['mean']})
```

### Observation
The number of departments grouped by are 12 for both the main and the subset Dataframe. Whereas, the key difference in the results comes in the mean of the order numbers for each department.

```python
# Creating new column "max_order"
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head()
```

```python
# Allocation of loyalty flags
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Verifying the allocation of loyalty flags
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Check Loyalty flag in dataframe
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number','max_order']].head(60)
```

```python
# Aggregating the min, max and mean of each loyalty flag
ords_prods_merge.groupby(['loyalty_flag']).agg({'prices':['mean','min','max']})
```

```python
# Creating column for average spender
ords_prods_merge['average_spend']=ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Creating flag for low and high spender in the dataframe
ords_prods_merge.loc[ords_prods_merge['average_spend']<10,'spender_flag']='Low_spender'
ords_prods_merge.loc[ords_prods_merge['average_spend']>= 10, 'spender_flag'] = 'High_spender'
```

```python
# Verify the  spender flags
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

```python
# Generate a column for frequent users
ords_prods_merge['Customer_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Creating flag on order frequency
ords_prods_merge.loc[ords_prods_merge['Customer_frequency'] > 20, 'frequency_flag'] = 'Non_frequent_customer'
ords_prods_merge.loc[(ords_prods_merge['Customer_frequency']>10) & (ords_prods_merge['Customer_frequency']<=20),'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['Customer_frequency'] <= 10, 'frequency_flag'] = 'Frequent Customer'
```

```python
# Verify the frequency flags
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
